# 3 different pipline analysis of the dataset related to data imputation

```python
import numpy as np
import pandas as pd
import statistics
from knn_impute import *
import matplotlib.pyplot as plt
from impyute.imputation.cs import fast_knn
from sklearn.metrics import roc_auc_score, roc_curve, f1_score, recall_score, average_precision_score
from matplotlib.colors import ListedColormap
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import StratifiedKFold, train_test_split, cross_val_score
from sklearn.preprocessing import OneHotEncoder, LabelEncoder
from sklearn.feature_selection import f_classif
```

### loading the dataset and changing categorical data into decimal form

```python
df = pd.read_csv('dataset.txt', encoding='ISO-8859-1', delimiter='\t', dtype=str)
X = df.iloc[:, 2:13]
a,b='',''
labels = X.columns
X = np.array(X)
for i in range(len(X)):
    for j in range(len(X[i])):
        if type(X[i,j]) == float: continue
        if X[i,j][0] == 'A': a = '00'
        if X[i,j][0] == 'T': a = '01'
        if X[i,j][0] == 'C': a = '10'
        if X[i,j][0] == 'G': a = '11'
        if X[i,j][2] == 'A': b = '00'
        if X[i,j][2] == 'T': b = '01'
        if X[i,j][2] == 'C': b = '10'
        if X[i,j][2] == 'G': b = '11'
        dec = int(a+b, 2)
        X[i, j] = int(a+b, 2)
X = X.astype(np.float)
imputed_X=fast_knn(X, k=30)
y = df.iloc[:, 0]
label_encoder = LabelEncoder()
y = label_encoder.fit_transform(y)
for i in range(len(y)):
    if y[i] == 1: y[i] = 0
    else: y[i] = 1
lr_clf = LogisticRegression(solver='lbfgs', max_iter=1000)
rf_clf = RandomForestClassifier(n_estimators=1000)
svm_clf = SVC(gamma='auto', probability=True)
```

```python
def plotScores(my_list, title_10fold, title_5fold, filename):
    labels = ['LR', 'RF', 'SVM']
    titles = [title_10fold, title_5fold]
    plt.figure(figsize=(12, 8))
    index = np.arange(len(labels))
    colors = ('green', 'blue', 'orange', 'red')
    for i in range(1, 3):
        ax1 = plt.subplot(2, 2, i)
        plotter = ax1.bar(index, my_list[i-1], color=colors)
        ax1.title.set_text(titles[i-1])
        ax1.legend((plotter[0], plotter[1], plotter[2]), my_list[i-1], loc=4)
        ax1.set_ylim((0, 100))
        ax1.set_xlabel('Classifiers', fontsize=10)
        ax1.set_ylabel('Classifiers\'s Scores', fontsize=10)
        ax1.set_xticks(index)
        ax1.set_xticklabels(labels, fontsize=11, rotation=30)
    plt.subplots_adjust(hspace=0.7, wspace=0.4)
    plt.savefig(filename, dpi=300, bbox_inches='tight')
    plt.show()
```

```python
def find_f1(clf, X, y, n):
    scores = cross_val_score(clf, X, y, cv=n, scoring='f1')
    return round(scores.mean()*100,2)
```

```python
def plot_features_table(clf, X_train, y_train, title, features, filename, y_pad=1.5):
    clf.fit(X_train, y_train)
    importances = rf_clf.feature_importances_
    indices = np.argsort(importances)[::-1]
    cell_text = np.array([[round(i,5)] for i in importances[indices]])
    collabel=['feature_importance_score']
    plt.title('Feature Importance Scores ({})'.format(title), y=y_pad, x=0.4, loc='center')
    plt.axis('off')
    the_table = plt.table(cellText=cell_text, rowLabels=features[indices], colLabels=collabel, loc='center', colWidths=[0.5], cellLoc='center')
    plt.savefig(filename, dpi=300, bbox_inches='tight')
    plt.show()
```

```python
def plot_f_p_table(data, title, features, filename, y_pad=1.5):
    f_values, p_values = f_classif(data, y)
    indices = np.argsort(f_values)[::-1]
    collabel = ['f-values', 'p-values']
    cell_text = np.column_stack((np.round(f_values[indices],5), np.round(p_values[indices],10)))
    plt.title('F and P Scores ({})'.format(title), y=y_pad, x=0.4, loc='center')
    plt.axis('off')
    the_table = plt.table(cellText=cell_text, rowLabels=features[indices], colLabels=collabel, loc='center', cellLoc='center')
    plt.savefig(filename, dpi=300, bbox_inches='tight')
    plt.show()
```

### approach A: applying knn-imputation after one-hot encoding

```python
X_optA = pd.get_dummies(df.iloc[:, 2:13])
optA_features = X_optA.columns
for feature_name in labels:
    X_optA.loc[df.iloc[:, 2:13][feature_name].isnull(), X_optA.columns.str.startswith(feature_name+'_')] = np.nan
X_optA = np.array(X_optA)
imputed_X_optA=np.rint(fast_knn(X_optA, k=30))
f1_10fold_title = '10-Fold f1-scores (knn-imputation after one-hot encoding)'
f1_5fold_title = '5-Fold f1-scores (knn-imputation after one-hot encoding)'
tenfold_f1scores = [find_f1(lr_clf, imputed_X_optA, y, 10), find_f1(rf_clf, imputed_X_optA, y, 10), find_f1(svm_clf, imputed_X_optA, y, 10)]
fivefold_f1scores = [find_f1(lr_clf, imputed_X_optA, y, 5), find_f1(rf_clf, imputed_X_optA, y, 5), find_f1(svm_clf, imputed_X_optA, y, 5)]
plotScores([tenfold_f1scores, fivefold_f1scores], f1_10fold_title, f1_5fold_title, 'f1 scores (option A)')
```

### feature importance f and p value tables of approach A

```python
X_train, X_test, y_train, y_test = train_test_split(imputed_X_optA, y, test_size=0.3, stratify=y, random_state=40)
plot_features_table(rf_clf, X_train, y_train, 'option A:\nknn-imputation after one-hot encoding', optA_features, 'option A feature importance scores')
plot_f_p_table(imputed_X_optA, 'option A:\nknn-imputation after one-hot encoding', optA_features, 'option A (f and p values)')
```
